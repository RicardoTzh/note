## dataframe的loc和iloc

loc和iloc区别在于，loc是行标签使用数字，iloc为列上使用数字

```python
df.loc[:,['省市','地市']]
```

```python
df.loc[:3,['省市','地市']]
```

```python
df.loc[[3,28],['省市','地市']]
```

```python
df.loc[df['地市']=='重庆市区']
```

```python
df.columns
```

```python
df.iloc[:,[3,7]]
```
